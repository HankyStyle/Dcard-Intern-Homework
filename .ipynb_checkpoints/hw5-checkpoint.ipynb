```python
import re
def remove_special(line):
    url = re.compile(r'[http|https]*://[a-zA-Z0-9.?/&=:]*',re.S)
    EOL = re.compile(r'\n', re.S)
    Sent_from = re.compile(r'Sent [from|From](.)*--', re.S)
    qoute = re.compile(r'※ 引述《(.)*》之銘言', re.S)
    line = re.sub(url, '',line)
    line = re.sub(EOL, '',line)
    line = re.sub(Sent_from, '',line)
    line = re.sub(qoute, '',line)
    return line
```

```python
all_data=[]
each_dict={}
content=""
title=""
id=""
for each_passage in data['hits']:
  each_dict.clear()
  id=str(each_passage['_source']['article_id'])
  content=remove_special(each_passage['_source']['content'])
  title=each_passage['_source']['article_title']
  each_dict.update({"id":id})
  each_dict.update({"title":title})
  each_dict.update({"contents":content})
  all_data.append(each_dict.copy())
with open("/content/corpus/Wiki.json" ,"w") as output:
  json.dump(all_data,output,indent=4)
```

```python
# 使用字典來統計每個字出現的次數
count_dict = {}
for word in collector:
    count_dict[word] = count_dict.get(word, 0) + 1

# 列印每個字出現的次數
for key, value in count_dict.items():
    print(f"{key} 出現了 {value} 次")
```

```python
freq = ""
for key, value in count_dict.items():
    freq += (f"'{key}': {value},")
```
